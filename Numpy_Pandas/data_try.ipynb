# PANDAS

```python
import pandas as pd

# Read a CSV file into a DataFrame
df = pd.read_csv('vgsales.csv')

# Display the first few rows of the DataFrame
print("First 5 rows of the DataFrame:")
print(df.head())

# Basic DataFrame Information
print("\nDataFrame Info:")
print(df.info())

# Descriptive Statistics
print("\nDescriptive Statistics:")
print(df.describe())

# Selecting Columns
selected_columns = df[['Name', 'Age']]
print("\nSelected Columns:")
print(selected_columns.head())

# Filtering Rows
filtered_rows = df[df['Age'] > 25]
print("\nFiltered Rows (Age > 25):")
print(filtered_rows.head())

# Grouping and Aggregation
grouped_data = df.groupby('Gender').agg({'Age': 'mean', 'Salary': 'sum'})
print("\nGrouped and Aggregated Data:")
print(grouped_data)

# Adding a New Column
df['Income_Category'] = pd.cut(df['Salary'], bins=[0, 50000, 100000, float('inf')], labels=['Low', 'Medium', 'High'])
print("\nDataFrame with New Column:")
print(df.head())

# Handling Missing Values
df.dropna(inplace=True)
print("\nDataFrame after Dropping Missing Values:")
print(df.head())

# Saving DataFrame to a New CSV file
df.to_csv('modified_example.csv', index=False)
print("\nDataFrame Saved to 'modified_example.csv'")
```

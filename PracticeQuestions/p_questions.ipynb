1. Given an array nums. We define a running sum of an array as runningSum[i] = sum(nums[0]…nums[i]).

   Return the running sum of nums.

```python
n = [1, 2, 6 , 4, 5]

for i in range(1, len(n)):
    n[i] += n[i - 1]
print(n)
```

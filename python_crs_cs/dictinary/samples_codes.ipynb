```python
def addTwoNumbers(l1 = [], l2 = []):
    # l1.reverse()
    # l2.reverse()
    num1 = ''
    num2 = ''
    for i in l1:
        num1 += str(i)
    for i in l2:
        num2 += str(i)
    num1 = int(num1)
    num2 = int(num2)
    total = num1 + num2
    total = str(total)
    r_list = []
    for i in total:
        val = int(i)
        r_list.append(val)
    f_list = r_list.reverse()

    return f_list

addTwoNumbers([2,3,4], [1,2,3])
```

```python
def addTwoNumbers(l1, l2):
    l1.reverse()
    l2.reverse()
    num1 = ''
    num2 = ''
    for i in l1:
        num1 += str(i)
    for i in l2:
        num2 += str(i)
    num1 = int(num1)
    num2 = int(num2)
    total = num1 + num2
    total = str(total)
    r_list = []
    for i in total:
        val = int(i)
        r_list.append(val)

    r_list.reverse()
    return r_list

addTwoNumbers([2],[2])
```

```python
# trying try-except block

print('give two numbers I will add it and give you the result')
print('Write "q" to quit.')

while True:
    first_number = input('\nFirst number: ')
    if first_number == 'q':
        break
    second_number = input('Second number: ')
    try:
        result = int(first_number) / int(second_number)
    except ZeroDivisionError:
        print('You can not divide by zero!')
    else:
        print(f'{result:.2f}')
```

```python
import json

numbers = [11,  7, 200, 2, 20]
filename = 'numbers.json'

with open(filename, 'w') as f_obj:
    json.dump(numbers, f_obj)

with open(filename) as f_obj:
    numbers = json.load(f_obj)

print(numbers)
```

```python
# testing a code by using a unittest module

def fullname(first_name, second_name):
    full_name = f'{first_name} {second_name}'
    return full_name.title()


import unittest

class NameTestCase(unittest.TestCase):
    def test_names(self):
        formated_name = fullname('John', 'Python')
        self.assertEqual(formated_name, 'Jonh Python')

# unittest.main()
```
